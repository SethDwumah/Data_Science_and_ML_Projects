## Aggregation

```python
## Aggregating information at "Embarked" column level
ps.sqldf("""
Select 
    df.Embarked,
    AVG(df.Age) As avg_age,
    Sum(df.Age) as total_age,
    Min(df.Age) as min_age,
    Max(df.Age) as Max_age,
    Count(df.PassengerId) as total_passengers,
    Count(Distinct df.Ticket) as dist_tickets
From titanic_dataset as df
Group by df.Embarked
 


""")
```

```python
ps.sqldf("""

Select 
    df.*
from titanic_dataset as df
limit 10


""")
```

```python
# Aggregating information at "Embarked " and "Sex" level
ps.sqldf("""
Select 
    df.Embarked,
    df.Sex,
    Avg(df.age) as avg_age,
    Sum(df.age) as total_age,
    Min(df.age )as min_age,
    Max(df.age) as max_age,
    Count(df.PassengerId) as num_passengers,
    Count(Distinct df.Ticket) as dist_tickets
from titanic_dataset as df
Group by df.Embarked,
    df.Sex




""")
```

```python
#and also using "where" to filter basis some conditions
ps.sqldf("""
Select
    df.Embarked,
    df.Sex,
    Avg(df.Age) as avg_age,
    Sum(df.age) as total_age,
    Min(df.age ) as min_age,
    Max(df.age) as max_age,
    Count(df.PassengerId) as tot_passengers,
    Count(Distinct df.Ticket) as dist_tickets
from titanic_dataset as df

where age >=18 and age<= 60

Group by df.Embarked,
    df.Sex


""")
```

```python
#Aggregating information at "Embarked" and "Sex" levels 
#and also using "WHERE" to filter basis some conditions
#and also sorting the output in descending order
ps.sqldf("""

SELECT 
    df.Embarked,
    df.Sex,
    AVG(df.Age) AS avg_age,
    SUM(df.Age) AS tot_age,
    MIN(df.Age) AS min_age,
    MAX(df.Age) AS max_age,
    COUNT(df.PassengerId) AS tot_passengers,
    COUNT(DISTINCT df.Ticket) AS dist_tickets
    
FROM titanic_dataset AS df

WHERE age >= 18
    AND age <=60
    
GROUP BY df.Embarked,
    df.Sex

ORDER BY df.Embarked DESC,
    df.Sex DESC

""")
```

```python
## Using "Having" to filter table
ps.sqldf("""
Select
    df.Embarked,
    df.Sex,
    AVG(df.age) as avg_age,
    Sum(df.age) as total_age,
    Min(df.age ) as min_age,
    Max(df.age) as max_age,
    Count(df.PassengerId) as total_passengers,
    Count(Distinct df.Ticket) as dist_tickets
From titanic_dataset as df

Where age>=18 and age<=60

Group by df.Embarked,
        df.Sex
Having avg_age>30
Order by df.Embarked DESC,
        df.Sex Desc




""")
```

## Joins

```python
cities = pd.DataFrame({'code':['S', 'C','L'],'city':['Southampton','Cherbourg','London']})
cities
```

```python
##inner join
ps.sqldf("""
Select 
    c.city,
    df.Embarked,
    df.Sex,
    Avg(df.age) as avg_age,
    Sum(df.age) as tot_age,
    Min(df.age) as min_age,
    Max(df.age) as max_age,
    Count(df.PassengerId) as tot_passengers,
    Count(Distinct df.ticket) as dist_tickets
From titanic_dataset as df

Inner Join cities as c
    on df.Embarked =c.code
Group by df.Embarked, df.Sex



""")
```

```python

##inner join
ps.sqldf("""
Select 
    c.city,
    df.Embarked,
    df.Sex,
    Avg(df.age) as avg_age,
    Sum(df.age) as tot_age,
    Min(df.age) as min_age,
    Max(df.age) as max_age,
    Count(df.PassengerId) as tot_passengers,
    Count(Distinct df.ticket) as dist_tickets
From titanic_dataset as df

Left Join cities as c
    on df.Embarked =c.code
Group by df.Embarked, df.Sex



""")
```

```python
#Union
ps.sqldf("""
Select Distinct df.Embarked from titanic_dataset as df
 union
 Select Distinct c.code from cities as c



""")
```

```python
## Using Rank()
ps.sqldf("""
Select
    df.Name,
    df.Embarked,
    df.Sex,
    df.Age,
    df.fare,
    Rank() Over(Order by df.Age DESC) as rnk
from titanic_dataset as df



""")
```
