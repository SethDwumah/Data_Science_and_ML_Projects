### Check NaN

```python
data.isna().sum()
```

```python
# replace Nan values with mean

data['azimuth'] = data['azimuth'].replace(np.NaN,data['azimuth'].mean())

data['tvd (ft)'] = data['tvd (ft)'].replace(np.NaN, data['tvd (ft)'].mean())
data['well spacing'] = data['well spacing'].replace(np.NaN, data['well spacing'].mean())
data['porosity'] = data['porosity'].replace(np.NaN, data['porosity'].mean())
data['water saturation'] = data['water saturation'].replace(np.NaN,data['water saturation'].mean())
data['toc'] = data['toc'].replace(np.NaN, data['toc'].mean())
data['youngs modulus'] = data['youngs modulus'].replace(np.NaN,data['youngs modulus'].mean())
data['isip']  = data['isip'].replace(np.NaN,data['isip'].mean())
data['breakdown pressure'] = data['breakdown pressure'].replace(np.NaN,data['breakdown pressure'].mean())
data['proppant volume'] = data['proppant volume'].replace(np.NaN,data['proppant volume'].mean())
```

```python
data.isna().sum()
```

### Descriptive Statistics

```python
numerical_attributes = data.select_dtypes(include=['int64', 'float64'])
categorical_attributes = data.select_dtypes(exclude=['int64','float64'])
```

```python
# Central Tendency: mean, median
ct1 = pd.DataFrame(numerical_attributes.apply(np.mean)).T
ct2 = pd.DataFrame(numerical_attributes.apply(np.median)).T



# Dispersion: std, min, max, range, skew , kurtosis

d1 = pd.DataFrame(numerical_attributes.apply(np.std)).T
d2 = pd.DataFrame(numerical_attributes.apply(min)).T
d3 = pd.DataFrame(numerical_attributes.apply(max)).T 
d4 = pd.DataFrame(numerical_attributes.apply(lambda x : x.max()-x.min())).T
d5 = pd.DataFrame(numerical_attributes.apply(lambda x: x.skew())).T
d6 = pd.DataFrame(numerical_attributes.apply(lambda x : x.kurtosis())).T

# Concantenate

m = pd.concat([d2,d3,d4,ct1,ct2,d1,d5,d6]).T.reset_index()
m.columns = ['attributes','min','max','range','mean','median','std','skew','kurtosis']
```
