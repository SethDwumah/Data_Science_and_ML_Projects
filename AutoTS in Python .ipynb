# AutoTS in Python

## Using AutoTS Library to predict stock price

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Apple_data.csv')
```

```python
df.head()
```

## Visualize the dataset 

```python
df.plot(subplots=True,figsize=(15,8), grid=True)
plt.show()
```

## Lets visualize the target, Close column

```python
df = df[['Date','Close']]
#df['Date'] = pd.to_datetime(df.Date)

plt.figure(figsize=(12,8))

plt.style.use('fivethirtyeight')

plt.plot(df['Date'], df['Close'], label='Close Price')
plt.title("Apple Stock Price", fontsize=20)
plt.grid()
plt.legend()
plt.show()
```

# Using AutoTS Library 

## For automatic time series forecasting

```python
from autots import AutoTS
```

```python
df['Date'] = pd.to_datetime(df.Date,utc=True)
```

```python
model = AutoTS(forecast_length=10, frequency='infer',
              ensemble='simple', drop_data_older_than_periods=200)

model = model.fit(df,date_col='Date',value_col='Close', id_col=None)
```

So in the output, we will get the closing stock prices of Apple for the next 10 days.

So let's use the predict function and have a look at the output:

```python
prediction = model.predict()
forecast = prediction.forecast

print("Stock Price Prediction of Apple")

print(forecast)
```
