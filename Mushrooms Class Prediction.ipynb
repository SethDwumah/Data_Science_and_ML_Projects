### Problem Statement

You are given dataset of mushrooms.The mushrooms dataset contain mushrooms of different kind and characteristics.From the dataset Class of each mushroom is stated whether Edible or Poisonous.

Use RandomForest and DecisionTree algorithm to predict the class of the mushroom. 

```python
#Import libraries
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report


from sklearn.model_selection import train_test_split
```

```python
#load dataset

mushroom_data = pd.read_csv('mushrooms.csv')
mushroom_data.head()
```

### Preprocessing  -- Convert categorical variables to numerical variables

```python
mushroom_code = mushroom_data.copy()
```

```python
mushroom_code['cap-shape'] = mushroom_code['cap-shape'].astype('category').cat.codes
mushroom_code['cap-color'] = mushroom_code['cap-color'].astype('category').cat.codes
mushroom_code['cap-surface'] = mushroom_code['cap-surface'].astype('category').cat.codes
mushroom_code['habitat'] = mushroom_code['habitat'].astype('category').cat.codes
mushroom_code['population'] = mushroom_code['population'].astype('category').cat.codes
mushroom_code['spore-print-color'] = mushroom_code['spore-print-color'].astype('category').cat.codes
mushroom_code['ring-type'] = mushroom_code['ring-type'].astype('category').cat.codes
mushroom_code['ring-number'] = mushroom_code['ring-number'].astype('category').cat.codes
mushroom_code['veil-color'] = mushroom_code['veil-color'].astype('category').cat.codes
mushroom_code['veil-type'] = mushroom_code['veil-type'].astype('category').cat.codes
mushroom_code['stalk-color-below-ring'] = mushroom_code['stalk-color-below-ring'].astype('category').cat.codes
mushroom_code['stalk-color-above-ring'] = mushroom_code['stalk-color-above-ring'].astype('category').cat.codes
mushroom_code['stalk-surface-below-ring'] = mushroom_code['stalk-surface-below-ring'].astype('category').cat.codes
mushroom_code['stalk-surface-above-ring'] = mushroom_code['stalk-surface-above-ring'].astype('category').cat.codes
mushroom_code['stalk-root'] = mushroom_code['stalk-root'].astype('category').cat.codes
mushroom_code['stalk-shape'] = mushroom_code['stalk-shape'].astype('category').cat.codes

mushroom_code['gill-color'] = mushroom_code['gill-color'].astype('category').cat.codes
mushroom_code['gill-size'] = mushroom_code['gill-size'].astype('category').cat.codes
mushroom_code['gill-spacing'] = mushroom_code['gill-spacing'].astype('category').cat.codes
mushroom_code['gill-attachment'] = mushroom_code['gill-attachment'].astype('category').cat.codes
mushroom_code['odor'] = mushroom_code['odor'].astype('category').cat.codes
mushroom_code['bruises'] = mushroom_code['bruises'].astype('category').cat.codes
mushroom_code['cap-surface'] = mushroom_code['cap-surface'].astype('category').cat.codes



mushroom_code['class'] = mushroom_code['class'].astype('category').cat.codes
```

```python
mushroom_code.head()
```

```python
mushroom_code.info()
```

### Set feature and target set from the dataset

```python
#feature
X = mushroom_code.iloc[:,1:]
#target
Y = mushroom_code.iloc[:,:1]
```

```python
X.head()
```

### Build the DecisionTree model

```python
# Split the dataset into train and test set
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.2, random_state=365)
```

```python
# clf = DecisionTreeClassifier

clf = DecisionTreeClassifier(random_state=365)
```

```python
# Train the model
clf.fit(x_train,y_train)
```

```python
# Predict the target
y_pred = clf.predict(x_test)
```

```python
y_pred[:5]
```

```python
print(classification_report(y_test,y_pred))
```

### RandomForest model


```python
RF = RandomForestClassifier(random_state=365)
```

```python
RF.fit(x_train,y_train)
```

```python
pred = RF.predict(x_test)
```

```python
print(classification_report(y_test,pred))
```
