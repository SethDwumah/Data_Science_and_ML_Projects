```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
from datetime import date, timedelta
import plotly.express as px
import yfinance as yf
```

```python
today = date.today()
```

```python
d1 = today.strftime("%Y-%m-%d")
```

```python
end_date = d1
```

```python
d2 = date.today()- timedelta(days=720)
```

```python
d2 = d2.strftime("%Y-%m-%d")
```

```python
start_date = d2
```

```python
data = yf.download('AAPL', start=start_date,end=end_date, progress=False)
```

```python
data.head()
```

```python
figure = px.line(data, x =data.index, y='Close', title="Time Series Analysis")
figure.show()
```

```python
import plotly.graph_objects as go

fig = go.Figure(data=[go.Candlestick(x=data.index, open = data['Open'],
                                    high=data['High'],
                                    low=data['Low'],
                                    close=data['Close'])])
fig.update_layout(title='Time Series Analysis (Candlestick Chart)',
                 xaxis_rangeslider_visible=False)

fig.show()

```

```python
fig = px.bar(data, x=data.index, y="Close",
            title="Time Series Analysis")
fig.show()
```


### Analyze stock price between the period of two specific dates

```python
fig = px.line(data, x=data.index, y='Close',
             range_x = ['2022-06-1','2022-12-31'],
             title='Time Series Analysis (Custome Date)')
fig.show()
```

## Interactive Candlestick 

```python
fig = go.Figure(data=[go.Candlestick(x=data.index,
                                    open=data["Open"],
                                    high=data["High"],
                                    low =data['Low'],
                                    close=data["Close"])])
fig.update_layout(title="Time Series Analysis (Candlestick)")

fig.update_xaxes(rangeslider_visible=True,
                 rangeselector = dict(
                 buttons=list([
                     dict(count =1, label='1m', step = 'month', stepmode='backward'),
                     dict(count =6 , label='6m', step='month',stepmode='backward'),
                     dict(count =1, label= 'YTD', step='year', stepmode='todate'),
                     dict(count=1, label='year', step='year', stepmode='backward'),
                     dict(step='all')
                 ])
                 )
                )
fig.show()
```
